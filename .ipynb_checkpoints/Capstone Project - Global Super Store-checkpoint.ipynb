# Data Preprocessing

```python
# Finding the null values in the dataset
df.isnull().sum()
```

```python
# Dropping the postal code column since it has more than 41000 missing values
df1 = df.drop('Postal Code', axis = 1)
```

```python
df1.isnull().sum()

# There is no null values present in df1
```

# Data types of the dataset

```python
# Datatypes of the columns
df1.dtypes
```

```python
# Changing the data type of features 'order date' and 'Ship date' to object
df1['Order Date'] = df1['Order Date'].astype(str)
df1['Ship Date'] = df1['Ship Date'].astype(str)
```

```python
# Categorical variables
df_cat = df1.select_dtypes(include = 'object')
df_cat.info()
```

```python
# Numerical variables variables
df_num = df1.select_dtypes(exclude = 'object')
df_num.info()
```

# Statistical summary of features

```python
# Statistical summary of numeric features 
df1.select_dtypes(exclude = 'object').describe().T
```

```python
# Statistical summary of categorical features
df1.select_dtypes(include = 'object').describe().T
```

# Redundency Reduction

```python
# Checking data for duplicate records
df1.nunique()
```

```python
# Removing the duplicate order id
df1.drop_duplicates(subset = 'Order ID' , inplace = True)
df1.shape
```

```python
df1.nunique()
```

```python
# Statistical summary of numeric features after removing the duplicate order id
df1.select_dtypes(exclude = 'object').describe().T
```

```python
# Statistical summary of categorical features after removing the duplicate order id
df1.select_dtypes(include = 'object').describe().T
```
